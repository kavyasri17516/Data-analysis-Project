```python
# load data
df=pd.read_csv("C:/Users/Admin/Desktop/CSV FILE(diversity inclusion).csv")
```

```python
# 1) Exploration
print(df.head())
```

```python
#2) Gender Distribution
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(data=df,x='Gender',hue='Gender',palette=['red','blue'])
plt.title("GENDER DISTRIBUTION")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#3)Employee by Department and Gender
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='Department @01.07.2020', hue='Gender',palette=['pink','o'])
plt.title("Employee by Department and Gender", fontsize=14, weight='bold')
plt.xlabel("Department", fontsize=12)
plt.ylabel("Employee Count", fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate Labels for readability
plt.legend(title="Gender")
plt.tight_layout()
plt.show()
```

```python
# 4)Years of Service vs Gender
plt.figure(figsize=(7,5))
sns.violinplot(data=df,x="Gender",y="Years since last hire",hue="Gender",palette=["Red","Orange"])
plt.xlabel("Gender")
plt.ylabel("Years of service")
plt.title("Years of service by gender")
plt.ylim(0)
plt.tight_layout()
plt.show()
```

```python
# 5)Correlation heatmap

import matplotlib.colors as mcolors

numeric_val=df.select_dtypes(include=["int64","float64"]).columns.tolist()
numeric_val.remove('Employee ID')
numeric_val.remove('Target hire balance')

corr=df[numeric_val].corr()

colors=["Blue","Red","Green","pink","orange","yellow"]
cmap=mcolors.ListedColormap(colors)

sns.heatmap(corr, annot=True,cmap=cmap, fmt=".2f", linewidths=1)
plt.title('Correlation Heatmap of Diversity and Inclusion of Employees',pad=30)
plt.xticks(rotation=270)
square=True
plt.show()
```
